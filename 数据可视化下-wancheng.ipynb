```python
import numpy as np
import csv
import pandas as pd
csv_files=r'D:\TJU\学习\GITHUB\ceshi.csv'
with open(csv_files,encoding='utf-8') as csvfile:
    read_CSV= csv.reader(csvfile)
    print(type(read_CSV))
    print(read_CSV[1])
    lines=csvfile.readlines()
    row=[]#定义行数组 
    for line in lines:
        row.append(line.split(','))
    column_x=[]#定义列数组
    column_y=[]#定义列数组
    column_z=[]#定义列数组
    for col in row:
        column_x.append(col[0])
    for col in row:
        column_y.append(col[1])
    for col in row:
        column_z.append(col[2])
```
